```python
import pandas as pd
```

```python
a = pd.read_csv("DS1.csv")
b = pd.read_csv("DS2.csv")
```

```python
d=a.rename(columns = {"name": "ship_name"})
```

```python
e=b.rename(columns = {"ship": "ship_name"})
```

```python
for i in d:
    if i not in e:
        for j in e:
            if j in i:
                f=e.rename(columns={j:i})
```

```python
final = pd.concat([d,f],ignore_index=True)
```
